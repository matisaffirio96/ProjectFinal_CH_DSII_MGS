```python
##importaciones de librerias
import pandas as pd
import numpy as np
import os
```

```python
##lectura del dataframe
def read_dataset():

    df = pd.DataFrame()

    ##carpeta del dataset
    folder = "datasets/"
    df = pd.read_csv("{}Loan.csv".format(folder), header=0, delimiter=';')

    return df


df = read_dataset()
```

```python
df["ApplicationDate"] = pd.to_datetime(df["ApplicationDate"], format='mixed')
df["MonthlyIncome"] = pd.to_numeric(df['MonthlyIncome'], errors='coerce')
df["TotalDebtToIncomeRatio"] = pd.to_numeric(df['TotalDebtToIncomeRatio'], errors='coerce')
```

```python
df['LoanApproved'] = df['LoanApproved'].map({0:'NO APROBADO', 1: 'APROBADO'})
```
